### Exercici 9 - Paraules Buides
Utilitzant l'exercici 8 arreglar el problema amb les paraules buides. \
Les paraules buides causen que el programa falli, ja que al buscar la lletra en la posicio 0, es a dir, la priemra posicio no existeix i causa que surti de les posicions existents en la paraula.

```python
lista = []
for i in range(3):
    try:
        lista.append(input("Digues una paraula: "))
    except ValueError:
        print("Error")
        
lletra = ""
while len(lletra) != 1:
    try:
        lletra = str(input("Digues una lletra"))
    except ValueError:
        print("Digues una unica lletra...")

print(lista)

mostrar = list(filter(lambda x: lletra in list(x)[:1], lista))
print(mostrar)
```
