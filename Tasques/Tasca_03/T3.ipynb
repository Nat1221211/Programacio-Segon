#### **Comprehensions**
 Les comprehensions són una manera d'estalviar linies al utilitzar un bucle for per a afegir un valor a una llista, diccionari, o algun altre tipus d'array.

##### ***Exemples***

```python
llista = [i * (i -1) for i in range(40) if i % 2 == 0 if i > 2 if i < 38]
print("List Comprehension")
print(llista)
```

```python
set = {i **5 for i in range(20) if i % 2 != 0}
print("Set Comprehension")
print(set)
```

```python
diccio = {k: k**5 for k in range(1, 80) if k % 4 == 0 if k % 7 == 0}
print("Diccionari Comprehension")
print(diccio)
```

### **Set (Conjunt)**
Un conjunt és un tipus de variable que elimina automaticament els duplicats, al igual que les anteriors es un tipus de llista, i pot guardar numerosos valors, el recibir dos valors iguals o més, elimina els repetits.


```python
set1 = {"Hola", "Que tal", "Gat", "Gos", "Gat", "Hola"}
print(set1) 
```

Els conjunts (set) tenen la caracteristica de poder mostrar les diferencies entre dues llistes, per ordre:
- Unio (simbol |) agrupa tots els elements de les dues llistes i no els repeteix.
- Intersecció (simbol &) Mostra els elements que es troben en ambdues llistes.
- Diferencia (simbol -) Mostra els elements que estan en la primera llista i no en la segona.
- Dif. Simetrica (simbol ^) Mostra els elements que estan en un dels dos pero no en els dos.

```python
set1 = {1, 2, 3, 4, 5, 6}
set2 = {2, 4, 6, 8, 10}
print(set1 | set2)
print(set1 & set2)
print(set1 - set2)
print(set1 ^ set2)
```

### Metodes Diccionaris
Son metodes de la variable de tipus diccionari, retornen els seus valors.
#### get(), items(), keys(), values()
Com indica cada un dels seus noms permeten respectivament:
- get() Retorna el valor d'una posicio del diccionari.
- items() Retorna de forma iterable cada una de les posicions incloent la clau i valor de cada posicio.
- keys() Retorna de forma iterable cada una de les claus del diccionari.
- values() Retorna de froma iterable cada un dels valors del diccionari.

```python
dicti = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5}
print("get()")
print(dicti.get("B"), "\n")
print("items()")
for i in dicti.items():
    print(i)
print("\nkeys()")
for i in dicti.keys():
    print(i)
print("\nvalues()")
for i in dicti.values():
    print(i)
```

### Random i Seed
Random afegeix diverses funcions que permeten generar valors aleatoris amb les seves caracteristiques. \
Seed es un metode del import random, que inicialitza el generador de nombres aleatoris, especifica el valor d'inici del random pertant el valor que generara sera sempre en el mateix ordre, i el primer valor sempre sera el mateix.

```python
import random
for i in range(2):
    print(random.randint(1, 1000))
print("\nSeed = 10")
random.seed(10)
for i in range(2):
    print(random.randint(1, 1000))
print("\nSeed = 2 dins del for")
for i in range(2):
    random.seed(2)
    print(random.randint(1, 1000))
```

### Exercici 5 - Diccionari i Preguntes
#### Creare un diccionari amb uns valors, i donare resposta a certes preguntes.

```python
def TotalIMitja(dictio):
    PTotal = 0
    PMitja = 0
    total = 0
    for k, v in dictio.items():
        PTotal += v["€"]*v["Qty"]
        total += v["Qty"]
    PMitja = PTotal / total
    return PTotal, PMitja

compra = {
    "Pomes": {"Qty": 5, "€": 0.42},
    "Peres": {"Qty": 3, "€": 0.66},
}

compra["Taronjes"]={"Qty": 4, "€": 0.65}
print(compra)
print(f"Les peres han costat: {compra['Peres']["Qty"]*compra['Peres']["€"]}€")
total = 0
for k, v in compra.items():
    total += v["Qty"]
print(f"Total fruites: {total}")
print(f"Fruita més cara: {max(compra.items(), key=lambda x: x[1]["€"])}")
PTotal, PMitja = TotalIMitja(compra)
print(f"Preu Total de la Compra: {PTotal}")
print(f"Preu mitja per fruita: {PMitja}")
```

### Exercici 8 - Filter i lambda
Utilitzant filter i lambda apartir de l'exercici 7b que nomes mostri les paraules que comencin amb la lletra indicada.

```python
lista = []
for i in range(3):
    try:
        lista.append(input("Digues una paraula: "))
    except ValueError:
        print("Error")
        
lletra = ""
while len(lletra) != 1:
    try:
        lletra = str(input("Digues una lletra"))
    except ValueError:
        print("Digues una unica lletra...")

print(lista)
mostrar = list(filter(lambda x: list(x)[0]==lletra, lista))
print(mostrar)
```

### Exercici 9 - Paraules Buides
Utilitzant l'exercici 8 arreglar el problema amb les paraules buides. \
Les paraules buides causen que el programa falli, ja que al buscar la lletra en la posicio 0, es a dir, la priemra posicio no existeix i causa que surti de les posicions existents en la paraula.

```python
lista = []
for i in range(3):
    try:
        lista.append(input("Digues una paraula: "))
    except ValueError:
        print("Error")
        
lletra = ""
while len(lletra) != 1:
    try:
        lletra = str(input("Digues una lletra"))
    except ValueError:
        print("Digues una unica lletra...")

print(lista)

mostrar = list(filter(lambda x: lletra in list(x)[:1], lista))
print(mostrar)
```
