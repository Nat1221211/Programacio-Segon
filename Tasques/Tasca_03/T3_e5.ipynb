### Exercici 5 - Diccionari i Preguntes
#### Creare un diccionari amb uns valors, i donare resposta a certes preguntes.

```python
def TotalIMitja(dictio):
    PTotal = 0
    PMitja = 0
    total = 0
    for k, v in dictio.items():
        PTotal += v["€"]*v["Qty"]
        total += v["Qty"]
    PMitja = PTotal / total
    return PTotal, PMitja

compra = {
    "Pomes": {"Qty": 5, "€": 0.42},
    "Peres": {"Qty": 3, "€": 0.66},
}

compra["Taronjes"]={"Qty": 4, "€": 0.65}
print(compra)
print(f"Les peres han costat: {compra['Peres']["Qty"]*compra['Peres']["€"]}€")
total = 0
for k, v in compra.items():
    total += v["Qty"]
print(f"Total fruites: {total}")
print(f"Fruita més cara: {max(compra.items(), key=lambda x: x[1]["€"])}")
PTotal, PMitja = TotalIMitja(compra)
print(f"Preu Total de la Compra: {PTotal}")
print(f"Preu mitja per fruita: {PMitja}")
```
