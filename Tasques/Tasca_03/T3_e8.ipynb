### Exercici 8 - Filter i lambda
Utilitzant filter i lambda apartir de l'exercici 7b que nomes mostri les paraules que comencin amb la lletra indicada.

```python
lista = []
for i in range(3):
    try:
        lista.append(input("Digues una paraula: "))
    except ValueError:
        print("Error")
        
lletra = ""
while len(lletra) != 1:
    try:
        lletra = str(input("Digues una lletra"))
    except ValueError:
        print("Digues una unica lletra...")

print(lista)
mostrar = list(filter(lambda x: list(x)[0]==lletra, lista))
print(mostrar)
```
