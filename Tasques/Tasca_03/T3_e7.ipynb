### Exercici 7 - Longituds de paraula
#### Creem tres programes que comptin la longitud de unes paraules, cada programa amb un metode diferent (for, map, comprehension).

```python
print(" - Variant For - ")

lista = []
for i in range(3):
    try:
        lista.append(input("Digues una paraula: "))
    except ValueError:
        print("Error")

lletres = []
print(lista)
for j in lista:
    lletres.append(len(j))
print(lletres)

print(" - Variant Map - ")

print(lista)

lletresMap = list(map(lambda x: len(x), lista))
print(lletresMap)

print(" - Variant List Comprehension - ")

print(lista)

lletresComprehension = [len(p) for p in lista]
print(lletresComprehension)
```
