### Exercici 2 - Trimestres
#### Creare una llista amb 4 llistes amb els 4 trimestres i els seus mesos i les mostrare.

```python
trimestres = {
    "Trimestre 1": ["Gener", "Febrer", "Març"],
    "Trimestre 2": ["Abril", "Maig", "Juny"],
    "Trimestre 3": ["Juliol", "Agost", "Septembre"],
    "Trimestre 4": ["Octubre", "Novembre", "Desembre"]
  }

for i, j in trimestres.items():
    print(f"{i}: {j}")
    print(f"Primer Més: {j[0]}\n")
```
