```python
import random

class EntityBase():
    # Atributs
    nom = ""
    health = int()
    attack = int()
    defense = int()
    speed = int()

    baseXP = int()

    # Metodes
    def __init__(self, nom, hp, atk, defs, spd, baseXP):
        self.nom = nom
        self.health = hp
        self.attack = atk
        self.defense = defs
        self.speed = spd
        self.baseXP = baseXP
    
class Entity():
    # Atributs
    base = None
    level = 1

    exp = 0
    nextlvl = 20

    MaxHP = int()
    CurHP = int()
    ATK = int()
    DEF = int()
    SPD = int()

    # Metodes
    def __init__(self, base, level):
        self.base = base
        self.level = level
        self.DefineStats()
        
    def DefineStats(self):
        self.MaxHP = (15 + (self.base.health / 50) * self.level)
        self.ATK = (5 + (self.base.attack / 100) * self.level)
        self.DEF = (5 + (self.base.defense / 100) * self.level)
        self.SPD = (5 + (self.base.speed / 100) * self.level)

    def LevelUP(self, target):
        xp = target.base.baseXP * 1 + (target.level / 10)
        self.exp += xp
        print(f"\n{self.base.nom} ha guanyat {xp} punts d'experiencia...")
        if self.nextlvl < self.exp:
            self.level+=1
            print(f"{self.base.nom} ha pujat de nivell, el seu HP s'ha recuperat Completament.")
            self.DefineStats()
            self.CurHP = self.MaxHP

            self.nextlvl += ((self.level**2) * 4) + 5

def GenerateBaseEntities():
    BaseEntities = (
        EntityBase("Player", 110, 110, 130, 120, 100),
        EntityBase("Dino", 100, 120, 100, 130, 28),
        EntityBase("Lando", 90, 100, 90, 110, 21),
        EntityBase("Kiyotaka", 80, 90, 80, 100, 20),
        EntityBase("Rem", 100, 110, 180, 90, 21),
        EntityBase("Arthur", 120, 100, 90, 120, 22),
        EntityBase("Black Star", 120, 140, 120, 150, 30),
        EntityBase("Maka", 90, 150, 90, 120, 22)
    )
    return BaseEntities

def GenerateEntities(BaseEntList):
    Ent = Entity(BaseEntList[random.randrange(1, len(BaseEntList))], 1)
    print(f"\n\n{Ent.base.nom} ha aparegut !! \n")
    return Ent

def BattleStart(Entity):
    Entity.CurHP = Entity.MaxHP

def Attack(Attacker, Target, power):
    print(f"\n\n{Attacker.base.nom} ha atacat !")
    damage = (((power * (Attacker.ATK / Target.DEF)) / 50) + 2) * (random.randrange(9, 12) / 10)
    
    Target.CurHP -= damage
    if Target.CurHP > 0:
        print(f"{Target.base.nom} ha recibit {round(damage, 2)} punts de dany, li queden {round(Target.CurHP, 2)} punts de vida.")
    if power >= 40:
        if power == 40:
            mal = damage / 4
        elif power == 60:
            mal = damage / 2.5
        Attacker.CurHP -= mal
        print("\nDegut a utilitzar un atac massa poderos s'ha recibit una part del mal.")
        print(f"{Attacker.base.nom} ha recibit {round(mal, 2)} punts de dany.")

def MostrarEstat(Entitat):
    print(f"{Entitat.base.nom}: {round(Entitat.CurHP, 2)} / {round(Entitat.MaxHP, 2)}   Level: {Entitat.level}")

def RegenerarEntitatEnDerrota(BaseEntities, Target, Attacker, state):
    jugadorderrotat = False
    if Attacker.CurHP <= 0:
        if Attacker.base.nom == "Player":
            print("Has estat Derrotat... \n")
            jugadorderrotat = True
    if Target.CurHP <= 0:
        if Target.base.nom == "Player":
            print("Has estat Derrotat... \n")
            jugadorderrotat = True
        elif jugadorderrotat == False:
            print(f"{Target.base.nom} ha estat derrotat.")
            print("Felicitats, has derrotat al enemic !!")
            Attacker.LevelUP(Target)
            Target = GenerateEntities(BaseEntities)
            BattleStart(Target)
            if Attacker.SPD >= Target.SPD:
                state["PlayerTurn"]=True
                state["COMTurn"]=False
            else:
                state["PlayerTurn"]=False
                state["COMTurn"]=True
    if jugadorderrotat == True:
        return Target, 5, state
    else:
        return Target, 0, state

def Fugir(Jugador, Enemic):
    probabilitatFugida = (0.2 + ((Jugador.SPD / Enemic.SPD) / 10) * (random.randrange(8, 13) / 10))
    if probabilitatFugida * 100 >= random.randrange(1, 101):
        return True
    else:
        return False

def Eleccio(QuantitatEleccions, TextError, AtacoEleccio):
    pos = 0
    while pos not in QuantitatEleccions:
        try:
            if AtacoEleccio == "Accio":
                print("\n1 -> Attacar")
                print("2 -> Fugir")
                pos = int(input("Que vols fer: "))
            elif AtacoEleccio == "Atac":
                print("\n1 -> Atac Normal")
                print("2 -> Atac Fort")
                print("3 -> Atac Demolidor")
                print("Quan més potent es l'atac mes consequencies tindra per a l'usuari...")
                pos = int(input("Quin atac vols fer: "))
        except ValueError:
            print(TextError)
    return pos
    
    
def main():
    pos = 0
    state = {"PlayerTurn": False, "COMTurn": False}
    playerAction = {"Attack": False}
    EnemyAction = {"Attack": False}
    BaseEntities = GenerateBaseEntities()
    PlayerEnt = Entity(BaseEntities[0], 1)


    print("\n ! Mini Joc ! \n")
    print("Presiona per a començar")
    input("")
    Enemy = GenerateEntities(BaseEntities)

    BattleStart(PlayerEnt)
    BattleStart(Enemy)

    if PlayerEnt.SPD >= Enemy.SPD:
        state["PlayerTurn"]=True
        state["COMTurn"]=False
    else:
        state["PlayerTurn"]=False
        state["COMTurn"]=True

    while pos != 5:
        pos = 0
        MostrarEstat(Enemy)
        MostrarEstat(PlayerEnt)
        if state["PlayerTurn"] == False:
            EnemyAction["Attack"]=True

            if EnemyAction["Attack"] == True:
                Attack(Enemy, PlayerEnt, random.randrange(1,41,20))
                PlayerEnt, pos, state = RegenerarEntitatEnDerrota(BaseEntities, PlayerEnt, Enemy, state)
            state["PlayerTurn"]=True
            state["COMTurn"]=False
        elif state["PlayerTurn"] == True:
            pos = Eleccio([1, 2], "Ha de dir un dels numeros per a la accio que fer...", "Accio")
            if pos == 1:
                pos = Eleccio([1, 2, 3], "Ha de dir un dels numeros per a l'atac que fer...", "Atac")
                playerAction["Attack"]=True
                if pos == 1:
                    power = 20
                elif pos == 2:
                    power = 40
                elif pos == 3:
                    power = 60
                Attack(PlayerEnt, Enemy, power)
                Enemy, pos, state = RegenerarEntitatEnDerrota(BaseEntities, Enemy, PlayerEnt, state)
                state["PlayerTurn"]=False
                state["COMTurn"]=True
            if pos == 2:
                succes = Fugir(PlayerEnt, Enemy)
                if succes == True:
                    print("Has fugit exitosament !!")
                    Enemy = GenerateEntities(BaseEntities)
                    BattleStart(Enemy)
                    if PlayerEnt.SPD >= Enemy.SPD:
                        state["PlayerTurn"]=True
                        state["COMTurn"]=False
                    else:
                        state["PlayerTurn"]=False
                        state["COMTurn"]=True
                else:
                    print(f"No has aconseguit fugir...")
                    state["PlayerTurn"]=False
                    state["COMTurn"]=True

    print("S'ha acabat la partida...")
        
    

if "__main__" == __name__:
    main()
```
