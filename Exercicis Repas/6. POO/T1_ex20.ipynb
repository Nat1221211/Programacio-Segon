###### Creem la classe Biblioteca així com la de Llibre, creem els objectes corresponents i probem els atributs i metodes. 

```python

class Llibre():
    # Atributs
    titol = ""
    autor = ""
    disponible = bool(True)

    # Metodes
    def __init__(self, titol, autor, disponible = True):
        self.titol = titol
        self.autor = autor
        self.disponible = disponible
    
    

class Biblioteca():
    # Atributs
    llibres = []
    nom = ""

    # Metodes
    def __init__(self, nom):
        self.nom = nom
    
    def AfegirLlibre(self):
        self.llibres.append(GenerarLlibre())

    def PrestarLlibre(self):
        prest = ""
        trobat = False
        for i in self.llibres:
            if prest == i.titol:
                trobat = True
        while trobat == False:
            try:
                prest = input("Digues el nom del llibre que vols agafar prestat: ")
                trobat = False
                for i in self.llibres:
                    if prest == i.titol:
                        trobat = True
                if trobat == False:
                    print("Has de dir el nom d'un dels llibres.")
            except ValueError:
                print("Has de dir el nom d'un dels llibres.")
        for i in self.llibres:
            if i.titol == prest:
                if i.disponible == True:
                    i.disponible = False
                else:
                    print("Aquest llibre ja ha estat prestat...")
    
    def RetornarLlibre(self):
        ret = ""
        trobat = False
        for i in self.llibres:
            if ret == i.titol:
                trobat = True
        while trobat == False:
            try:
                ret = input("Digues el nom del llibre que vols agafar prestat: ")
                trobat = False
                for i in self.llibres:
                    if ret == i.titol:
                        trobat = True
                if trobat == False:
                    print("Has de dir el nom d'un dels llibres.")
            except ValueError:
                print("Has de dir el nom d'un dels llibres.")
        for i in self.llibres:
            if i.titol == ret:
                if i.disponible == False:
                    i.disponible = True
                else:
                    print("Aquest llibre ja el tenim...")
    
    def MostrarLlibres(self):
        print("Tenim els llibres: ")
        for i in self.llibres:
            print(i.titol)

def GenerarLlibre():
        nom = ""
        while nom == "":
            try:
                nom = input("Digues el nom del llibre: ")
            except ValueError:
                print("Has de dir un nom.")
                nom = ""

        autor = ""
        while autor == "":
            try:
                autor = input("Digues el nom del autor del llibre: ")
            except ValueError:
                print("Has de dir un nom.")
                autor = ""
        res = ""
        while res not in ["F", "T"]:
            try:
                res = input("Esta el llibre disponible o no? (F per no / T per si)\n")
            except ValueError:
                print("Digues una resposta correcta.")
        if res == "F":
            disp = False
        else:
            disp = True
        llibre = Llibre(nom, autor, disp)
        return llibre

def GenerarBiblioteca():
    nom = ""
    while nom == "":
        try:
            nom = input("Digues el nom de la biblioteca: ")
        except ValueError:
            print("Has de dir un nom.")
            nom = ""
    quant = 0
    while quant <= 0:
        try:
            quant = int(input("Digues quants llibres té la teva biblioteca: "))
            if quant <= 0:
                print("Has de dir una quantitat de minim 1.")
        except ValueError:
            print("Has de dir una quantitat numerica.")
    biblioteca = Biblioteca(nom)
    for i in range(quant):
        biblioteca.AfegirLlibre()
    return biblioteca

def main():
    pos = 0
    biblioteca = None
    print("\n ! Biblioteca ! \n")
    print("1 - Afegir Llibre")
    print("2 - Prestar Llibre")
    print("3 - Retornar Llibre")
    print("4 - Mostrar Llista de llibres")
    print("5 - Sortir de la Biblioteca")
    while pos != 5:
        pos = 0
        while pos not in [1, 2, 3, 4, 5]:
            if biblioteca == None:
                print("Quina biblioteca vols utilitzar.")
                biblioteca = GenerarBiblioteca()
            else:
                try:
                    pos = int(input("Que vols fer: "))
                except ValueError:
                    print("Ha de dir un dels numeros per a la accio que fer...")

        if pos == 1:
            biblioteca.AfegirLlibre()
        elif pos == 2:
            biblioteca.PrestarLlibre()
        elif pos == 3:
            biblioteca.RetornarLlibre()
        elif pos == 4:
            biblioteca.MostrarLlibres()
        elif pos == 5:
            print("Has sortit de la biblioteca...")
    

if "__main__" == __name__:
    main()
```
