##### Exercici 18 - Classe Compte Bancari
###### Creem la classe Compte Bancari, creem l'objecte compte bancari desde la classe corresponent i probem els atributs i metodes. 

```python
import os

class Alumne():
    # Atributs
    titular = ""
    saldo = float()

    # Metodes
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = round(saldo, 2)
    
    def Ingresar(self, ingresar):
        self.saldo += ingresar
        self.saldo = round(self.saldo, 2)
        print(f"Has retirat {ingresar}€ del compte, queden {self.saldo}€.")

    def Retirar(self, retirar):
        self.saldo -= retirar
        self.saldo = round(self.saldo, 2)
        print(f"Has retirat {retirar}€ del compte, queden {self.saldo}€.")
    
    def MostrarSaldo(self):
        print(f"Saldo: {self.saldo}€")

def GenerarCompte():
    nom = ""
    while nom == "":
        try:
            nom = input("Digues el nom del titular del compte: ")
        except ValueError:
            nom = ""

    saldo = 0
    incorrecte = False
    while incorrecte != True:
        try:
            saldo = float(input("Digues el teu saldo: "))
            incorrecte = True
        except ValueError:
            incorrecte = False
    compte = Alumne(nom, saldo)
    return compte

def IngresRetir():
    valor = 0
    while valor <= 0:
        try:
            valor = float(input("Quantitat: "))
            if valor <= 0:
                print("Es necessita un valor positiu.")
        except ValueError:
            print("Valor incorrecte...")
    return valor

def main():
    pos = 0
    compte = None
    print("\n ! Compte Bancari ! \n")
    print("1 - Ingressar Saldo")
    print("2 - Retirar Saldo")
    print("3 - Revisar Saldo")
    print("4 - Sortir del Compte")
    while pos != 4:
        pos = 0
        while pos not in [1, 2, 3, 4]:
            if compte == None:
                print("No tens un compte, diguens la seguent informacio i et creem un.")
                compte = GenerarCompte()
            else:
                os.system("cls")
                pos = int(input("Que vols fer: "))

        if pos == 1:
            ingres = IngresRetir()
            compte.Ingresar(ingres)
        elif pos == 2:
            retir = IngresRetir()
            compte.Retirar(retir)
        elif pos == 3:
            compte.MostrarSaldo()
        elif pos == 4:
            print("Has sortit del compte...")
    

if "__main__" == __name__:
    main()
```
