##### Exercici 13 - Calculadora Modular
###### Creem quatre funcions una per a cada operacio basica (multiplicar, dividir, sumar, restar).

```python
import os

def sumar(actual):
    correcte = False
    while correcte == False:
        try:
            num = float(input("Digues el valor a sumar: "))
            correcte = True
        except ValueError:
            print("El caracter introduit no es un correcte.")
            correcte = False
    ant = actual
    actual += num
    print(f"{ant} + {num} = {actual}")
    return actual

def restar(actual):
    correcte = False
    while correcte == False:
        try:
            num = float(input("Digues el valor a restar: "))
            correcte = True
        except ValueError:
            print("El caracter introduit no es un correcte.")
            correcte = False
    ant = actual
    actual -= num
    print(f"{ant} - {num} = {actual}")
    return actual

def multiplicar(actual):
    correcte = False
    while correcte == False:
        try:
            num = float(input("Digues el multiplicador: "))
            correcte = True
        except ValueError:
            print("El caracter introduit no es un correcte.")
            correcte = False
    ant = actual
    actual *= num
    print(f"{ant} x {num} = {actual}")
    return actual

def dividir(actual):
    correcte = False
    while correcte == False:
        try:
            num = float(input("Digues el divisor: "))
            correcte = True
        except ValueError:
            print("El caracter introduit no es un correcte.")
            correcte = False
    ant = actual
    actual /= num
    print(f"{ant} / {num} = {actual}")
    return actual

def main():
    print("\n ! Calculadora Modular ! \n")
    
    print("Menu")
    print("+ -> Sumar")
    print("- -> Restar")
    print("x -> Multiplicar")
    print("/ -> Dividir")
    print("C -> Reiniciar Operacions")
    print("S -> Sortir de la calculadora")

    numero = 0
    char = "+"
    while char in ["+","-","x","/","C","S"]:
        char = ""
        while char not in ["+","-","x","/","C","S"]:
            os.system("cls")
            try:
                char = input("Digues que vols fer, selecciona el caracter corresponent: ")
                if char not in ["+","-","x","/","C","S"]:
                    print("El caracter introduit no coincideix amb una operació...")
            except ValueError:
                print("El caracter introduit no es un correcte.")
        print(f"Valor actual:  {numero}")
        if char == "+":
            numero = sumar(numero)
        elif char == "-":
            numero = restar(numero)
        elif char == "x":
            numero = multiplicar(numero)
        elif char == "/":
            numero = dividir(numero)
        elif char == "C":
             numero = 0
             print("Valor reiniciat")
        elif char == "S":
            print("Has sortit de la calculadora, programa Finalitzat...")
            break
    
        

if "__main__" == __name__:
    main()
```
